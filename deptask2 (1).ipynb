```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
e_data=pd.read_csv("/content/emails.csv")
e_data.head()
```

```python
clean_data=e_data.drop(columns=["Email No."])
clean_data.tail()
```

```python
clean_data.dropna(inplace=True)
```

```python
clean_data.isnull().sum()
```

```python
clean_data.describe()
```

```python
clean_data.info()
```

```python
plt.figure(figsize=(6,6),facecolor="red")
sns.countplot(x='Prediction',data=clean_data)
plt.title("Distribution of spam and not spam emails")
plt.xlabel("Spam 1 or not spam 0")
plt.ylabel("count")
plt.show()
```

```python
x=clean_data.drop(columns=['Prediction'])
y=clean_data['Prediction']
```

```python
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
naiveb=MultinomialNB()
naiveb.fit(x_train,y_train)
```

```python
y_pred_naiveb=naiveb.predict(x_test)
y_pred_naiveb
```

```python
sns.countplot(x=y_pred_naiveb, palette=['#FF6347', '#4682B4'])
plt.title("Predictions", fontsize=15)
plt.xlabel("Prediction", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
niaveb_acc=accuracy_score(y_test,y_pred_naiveb)
print("accuracy is :",niaveb_acc)
```

```python
niaveb_clasification=classification_report(y_test,y_pred_naiveb)
print("classifiaction is :",niaveb_clasification)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel='linear',random_state=43)
sv.fit(x_train,y_train)
```

```python
y_pred_sv=sv.predict(x_test)
y_pred_sv
```

```python
sns.countplot(x=y_pred_sv, palette=['#CC9994', '#8989B4'])
plt.title("Predictions", fontsize=15)
plt.xlabel("Prediction", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
sv_accuracy=accuracy_score(y_test,y_pred_sv)
print("accuracy is :",sv_accuracy)
```

```python
svm_classification=classification_report(y_test,y_pred_sv)
print("classification is :",svm_classification)
```
